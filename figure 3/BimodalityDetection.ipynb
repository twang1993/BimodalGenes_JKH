```python
def detect_bimodality(gene_abundance):
    gene_abundance = gene_abundance.values
    if len(gene_abundance) < 2:
        return False
    # Remove infinite and NaN values
    gene_abundance = gene_abundance[np.isfinite(gene_abundance)]
    if len(gene_abundance) < 2:
        return False
    
    # Perform PCA
    pca = PCA()
    gene_abundance_transformed = pca.fit_transform(gene_abundance.reshape(-1, 1))

    # Use the first principal component
    first_pc = gene_abundance_transformed[:, 0].reshape(-1, 1)

    # Fit unimodal and bimodal GMMs
    gmm_1 = GaussianMixture(n_components=1)
    gmm_2 = GaussianMixture(n_components=2)
    gmm_1.fit(first_pc)
    gmm_2.fit(first_pc)

    # Calculate BIC for both models
    bic_1 = gmm_1.bic(first_pc)
    bic_2 = gmm_2.bic(first_pc)

    # Decide based on BIC
    if bic_2 < bic_1:
        return True
    return False
```

```python
bimodal_genes = []
for i in range(len(KO_List)):
    if np.mod(i,100)==0:
        print(i)
    temp0=KO_AllThree.iloc[i,:]
    if detect_bimodality(np.log10(temp0)):
        bimodal_genes.append(KO_List[i])
```
