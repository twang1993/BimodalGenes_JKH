```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import gaussian_kde
```

```python
country_names=['3_countries_combined','American','Spanish','Danish']
dir1='genus normalization/'
dir2='bimodal genus/'
bimodalgenus=pd.read_csv(dir2+'3_countries_combined_genus_normalization_BiDgenus',index_col=0,sep='\t').index.tolist()
bimodalgenus.sort()
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
#plot bimodal distribution of genes
for j in range(len(country_names)):
    df=pd.read_csv(dir1+country_names[j]+'_genus_normalization',index_col=0,sep='\t')
    plt.figure(figsize=(11,1))
    for i in range(len(bimodalgenus)):
        plt.subplot(1,len(bimodalgenus),i+1)
        plt.subplots_adjust(wspace=0.05,hspace=0.05)
        y=df[bimodalgenus[i]]
        data = np.log10(y[y>0])
        sns.kdeplot(data,bw_adjust=1.0,fill=True,alpha=0.8,linewidth=0,color='#8A8AFF') 
        plt.ylabel('')
        plt.xlim(-9,1.0)
        plt.ylim(0,0.6)
        plt.tick_params(axis='both', labelsize=7,labelfontfamily='Arial')  
        if i == 0:
            plt.xticks(np.arange(-9,2.0,2))
            plt.yticks(np.arange(0,0.8,0.2))
        else:
            plt.xticks([]) 
            plt.yticks([]) 
        plt.savefig('fig/'+'7 bimodalgenus distribution in '+country_names[j] +'.pdf',bbox_inches='tight')
```
