```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import re
```

```python
# 设置字体为 TrueType 而不是矢量图形
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
# 全局设置字体为 Arial
plt.rcParams['font.family'] = 'Arial'

data = pd.read_csv('corration bimodal & all genus(sort)/3countriescombined_corration_bimodal&genus(class)',index_col=0,sep='\t')
df=pd.read_excel('genus taxonomy/'+'genus_taxonomy_all_sorted.xlsx')
class_list=df['class'].tolist()

first_occurrence_class = []
first_occurrence_class_index = []
# 遍历 class_list 列表
for i in range(len(class_list)):
    # 若元素还未在字典中，将其添加到字典，键为元素，值为索引
    if class_list[i] not in first_occurrence_class:
        first_occurrence_class.append(class_list[i])
        first_occurrence_class_index.append(i)
x_labels=first_occurrence_class
x_ticks=[x+0.5  for x in first_occurrence_class_index]
plt.figure(figsize=(20, 3))  # 设置图形大小
#sns.heatmap(data, annot=False, fmt=".2f", cmap="RdBu")  # 绘制热图
sns.heatmap(data, cbar_kws={'label': 'spearmanr\u2019\u03C1'}, cmap="bwr")  # 绘制热图("RdBu"：红色到蓝色的渐变;"coolwarm"：较柔和的蓝色到红色渐变;"bwr"：蓝色到白色到红色)#'\u2019':英文右边'；'\u03c1':符号‘ρ’。
plt.title('3 countries combined')  # 添加标题
plt.xticks(ticks=x_ticks,labels=x_labels,fontsize=4)
plt.savefig('fig/Heatmap for 11 bimodal genes & all genus(sort).pdf',bbox_inches='tight')
data.to_csv('fig/Heatmap for 11 bimodal genes & all genus(sort).csv',index=True,header=True,sep=',')
```
