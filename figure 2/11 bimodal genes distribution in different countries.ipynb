```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import gaussian_kde
```

```python
country_names=['3_countries_combined','American','Spanish','Danish']
bimodalgenes=pd.read_csv('bimodal genes/3_countries_combined_BiDgenes',index_col=0,sep='\t').index.tolist()
bimodalgenes.sort()
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
#plot bimodal distribution of genes
for j in range(len(country_names)):
    df=pd.read_csv('test2/'+country_names[j],index_col=0,sep='\t').T
    plt.figure(figsize=(11,1))
    for i in range(len(bimodalgenes)):
        plt.subplot(1,len(bimodalgenes),i+1)
        plt.subplots_adjust(wspace=0.05,hspace=0.05)
        y=df[bimodalgenes[i]]
        data = np.log10(y[y>0])
        sns.kdeplot(data,bw_adjust=1.0,fill=True,alpha=0.8,linewidth=0,color='#8A8AFF') 
        plt.ylabel('')
        plt.xlim(-9,-3)
        plt.ylim(0,0.5)
        plt.tick_params(axis='both', labelsize=7,labelfontfamily='Arial') 
        if i == 0:
            plt.xticks(np.arange(-9,-2,2))
            plt.yticks(np.arange(0,0.6,0.2))
        else:
            plt.xticks([])  
            plt.yticks([]) 
        plt.savefig('fig/'+'11 bimodalgenes distribution in '+country_names[j] +'.pdf')
```
