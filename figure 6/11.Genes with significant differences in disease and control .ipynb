```python
import matplotlib.pyplot as plt
import matplotlib.patches as pc
import pandas as pd
import numpy as np
from collections import Counter
disease_list = ['Arthritis high', 'Arthritis moderate', 'Colorectal cancer advanced adenoma','Colorectal cancer carcinoma', "Crohn's disease", 'Liver cirrhosis', 'Obesity', 'Type 2 diabetes', 'Ulcerative colitis']
gene_list = ['K02663', 'K03006', 'K03283', 'K03289', 'K07671', 'K09802', 'K11707', 'K11708', 'K11709', 'K11710', 'K12436']
dir2='relationship between disease and control peoples/'
#genes with significant difference in diseases and control.
significant_differences=pd.read_csv(dir2+'Diseases and control populations with significant differences',sep='\t')
significant_disease=significant_differences.iloc[:,0].tolist()
significant_genes=significant_differences.iloc[:,1].tolist()
significant_pvalue=significant_differences.iloc[:,2].tolist()
significant_pair=[]
for i in range(len(significant_disease)):
    if significant_pvalue[i]<0.05 :
        significant_pair.append([significant_disease[i],significant_genes[i]])
plt.figure(figsize=(2,3))
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 7
for i in range(len(disease_list)):
    for j in range(len(gene_list)):
        if [disease_list[i],gene_list[j]] in significant_pair:
            plt.plot(i,j,color='#C84741',alpha=0.8,marker='o',markersize=12, markeredgewidth=0)
#plt.tick_params(axis='x', top=True, bottom=False, labeltop=True, labelbottom=False)
plt.xticks(ticks=range(len(disease_list)),labels=disease_list,fontsize=7,rotation=45)
plt.yticks(ticks=range(len(gene_list)),labels=gene_list,fontsize=7)
plt.xlim(-0.8,len(disease_list))
plt.ylim(-0.8,len(gene_list))
plt.grid(True, linestyle='--', alpha=0.3)
plt.savefig('fig/genes with significant difference in diseases and control.pdf',bbox_inches='tight')      
```
