```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
dir0 = 'relationship between disease and control peoples/'
dir1 = 'gene abundance (disease)/'
gene_list=['K11707','K11708','K11709','K11710','K12436','K09802','K07671','K03289','K03283','K03006','K02663']
Disease_list=['Arthritis_High','Arthritis_Moderate','ColorectalCancer_AdvancedAdenoma','ColorectalCancer_Carcinoma','CrohnsDisease','LiverCirrhosis','Obesity','Type2Diabetes','UlcerativeColitis']
control_list=['Arthritis_None','Arthritis_None','ColorectalCancer_Controls','ColorectalCancer_Controls','CrohnsDisease_None','LiverCirrhosis_None','Obesity_None','Type2Diabetes_None','UlcerativeColitis_None']
Disease_list2=['Arthritis high','Arthritis moderate','Colorectal cancer advanced adenoma','Colorectal cancer carcinoma','Crohn\'s disease','Liver cirrhosis','Obesity','Type 2 diabetes','Ulcerative colitis']
P_Values=pd.DataFrame(np.zeros((len(Disease_list), len(gene_list))))
disease_clear=[]
p_value_clear=[]
gene_list_clear=[]
for i in range(len(Disease_list)):
    df0=pd.read_csv(dir1+control_list[i]+'_Gene_abundance',index_col=0,sep='\t')
    df1=pd.read_csv(dir1+Disease_list[i]+'_Gene_abundance',index_col=0,sep='\t')
    for j in range(len(gene_list)):
        data0=df0[gene_list[j]]
        data1=df1[gene_list[j]]
        # Test if the data follows a normal distribution
        statistic, p_value0 = stats.shapiro(data0)
        statistic, p_value1 = stats.shapiro(data1)
        if p_value0 > 0.05 and p_value1 > 0.05:
            print(Disease_list2[i]+"数据服从正态分布")
            _, p_value_align = stats.levene(data0, data1)
            if p_value_align > 0.05:
                # Normal distribution and equal variances, use independent samples t-test
                _, p_value = stats.ttest_ind(data0, data1, equal_var=True)
            else:
                # Normal distribution but unequal variances, use Welch's t-test
                _, p_value = stats.ttest_ind(data0, data1, equal_var=False)
        else:
            print(Disease_list2[i]+"数据不服从正态分布")
            # At least one column of data does not follow a normal distribution, use Mann - Whitney U test
            _, p_value = stats.mannwhitneyu(data0, data1)
        P_Values.iloc[i,j]=p_value
        if p_value < 0.05:
            Disease_clear_test=Disease_list2[i]
            gene_list_clear_test=gene_list[j]
            disease_clear.append(Disease_clear_test)
            gene_list_clear.append(gene_list_clear_test)
            p_value_clear.append(p_value)
        

P_Values.columns=gene_list
P_Values.index=Disease_list2
P_Values.to_csv('relationship between disease and control peoples/P values of disease and control peoples',sep='\t',index=True)

data_clear=pd.DataFrame({'Disease':disease_clear,'Gene':gene_list_clear,'P value':p_value_clear})
data_clear.to_csv('relationship between disease and control peoples/Diseases and control populations with significant differences',sep='\t',index=False)
```
