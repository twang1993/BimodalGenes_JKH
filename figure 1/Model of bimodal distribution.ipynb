```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import gaussian_kde
import random
```

```python
df=pd.read_csv('Gene_Abundances/Human_cheek',index_col=0,sep='\t').T
bimodalgenes=pd.read_csv('Bimodal_Genes/Human_cheek_BiDgenes',index_col=0,sep='\t').index.tolist()
y=df[bimodalgenes[15]]
data = np.log(y[y>0])
sns.kdeplot(data,bw_adjust=1.0,fill=True,color='#8A8AFF',alpha=1.0,linewidth=0) 
plt.ylabel('')
plt.xlabel('')
plt.xticks([])
plt.yticks([])
plt.savefig('fig/Modal of Biomodal Distribution.pdf')
```
