```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import random
import matplotlib
```

```python
# Set the font type to TrueType instead of vector graphics
plt.rcParams['pdf.fonttype'] = 42
# Set the global font to Arial
plt.rcParams['font.family'] = 'Arial'
microbiome_names=['3_countries_combined','American','Danish','Spanish']
for i in range(len(microbiome_names)):
    df1=pd.read_csv('Gene_Abundances/'+microbiome_names[i],index_col=0,sep='\t').T
    df2=pd.read_csv('Bimodal_Genes/'+microbiome_names[i]+'_BiDgenes',index_col=0,sep='\t')
    Bimodalgenes=df2.index.tolist()
    bigdata_genes=[gene for gene in df1.columns.tolist() if len([i for i in df1[gene] if i>0])>0.9*len(df1[gene])]
    Unimodalgenes=[gene for gene in bigdata_genes if gene not in Bimodalgenes]
    Example_Unimodalgenes=random.sample(Unimodalgenes,len(Bimodalgenes))
    
    # plot the bimodal distribution
    plt.figure(figsize=(4,len(Bimodalgenes)//4+1))
    for j in range(len(Bimodalgenes)):
        plt.subplot(len(Bimodalgenes)//4+1,4,j+1)
        plt.subplots_adjust(wspace=0.05,hspace=0.05)
        y=df1[Bimodalgenes[j]]
        data = np.log(y[y>0])
        sns.kdeplot(data,bw_adjust=1.0,color='#C62828') 
        plt.hist(data,density=True,bins=20,color='#8A8AFF')
        plt.ylabel('')
        plt.xlabel('')
        plt.xticks([])
        plt.yticks([])
        plt.legend(title=Bimodalgenes[j],fontsize=2.0,loc='upper left',frameon=False,)
    plt.savefig('fig'+'/'+microbiome_names[i]+'_Bimodal'+'.pdf')
    plt.close()
    # plot the unimodal distribution
    plt.figure(figsize=(4,len(Example_Unimodalgenes)//4+1))
    for k in range(len(Example_Unimodalgenes)):
        plt.subplot(len(Example_Unimodalgenes)//4+1,4,k+1)
        plt.subplots_adjust(wspace=0.05,hspace=0.05)
        y=df1[Example_Unimodalgenes[k]]
        data = np.log(y[y>0])
        sns.kdeplot(data,bw_adjust=1.0,color='#C62828') 
        plt.hist(data,density=True,bins=20,color='#8A8AFF')
        plt.ylabel('')
        plt.xlabel('')
        plt.xticks([])
        plt.yticks([])
        plt.legend(title=Example_Unimodalgenes[k],fontsize=2.0,loc='upper left',frameon=False)
    plt.savefig('fig'+'/'+microbiome_names[i]+'_Unimodal'+'.pdf')
    plt.close()
```
