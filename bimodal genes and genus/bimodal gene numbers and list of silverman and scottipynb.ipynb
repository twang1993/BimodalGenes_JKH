```python
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde
import os
```

```python
del_dir1='test'
del_dir2='bimodal genes'
os.makedirs(del_dir2,exist_ok=True) # If the folder del_dir2 does not exist, create a new one.
filelist=os.listdir(del_dir1) # Extract all file names in the folder as a list.
for file in filelist:
    del_file=del_dir1+'/'+file
    df=pd.read_csv(del_file,index_col=0,delimiter='\t') 
    dT=df.T
    #1.Finding Biomodal Distribution of genes
    Bimodal_genenames=[]
    Bimodal_genevalleys=[]
    for i in range(len(dT.columns)): # Write a loop based on the column length of the original data.
        y=dT.iloc[:,i]
        if np.sum(abs(np.log(y[y>0]))>0)>0.3*len(dT): # Write a loop based on the column length of the original data.
            data = np.log(y[y>0]) 
            #sns.kdeplot(data,bw_adjust=1) 
            kde = gaussian_kde(data,bw_method='scott') # Obtain the kernel density function result（scott）.
            kde_data = kde.evaluate(np.linspace(min(data), max(data), 1000))  
            kde_x=np.linspace(min(data), max(data), 1000) # Define the range of x values based on the minimum and maximum values of the data, and divide it into 1000 equal parts. 
            kde_y=kde(kde_x) #按x值求核密度函数的相应的y值
            dy_dx = np.gradient(kde_y, kde_x) # Differentiate the kernel density function.
            peakposi_y=[] # Y values of peaks
            valeyposi_y=[] # Y values of valleys
            peakposi_x=[] # X values of peaks
            valeyposi_x=[] # X values of valleys
            for j in range(len(dy_dx)):
                if dy_dx[j-1]>=0 and dy_dx[j]<=0 :   # Use the derivative equal to 0 as the screening condition for peaks and valleys.
                    peak_y=kde_y[j-1]
                    peakposi_y.append(peak_y)
                    peak_x=kde_x[j-1]
                    peakposi_x.append(peak_x)
                if dy_dx[j-1]<=0 and dy_dx[j]>=0:
                    valey_y=kde_y[j]
                    valeyposi_y.append(valey_y)
                    valey_x=kde_x[j]
                    valeyposi_x.append(valey_x)
            if len(peakposi_y)>=2: # Screen out those with more than two peaks
                Ymax1=max(peakposi_y) # Maximum peak value
                BB=[]
                for a in peakposi_y: 
                    BB.append(a)
                BB.remove(max(BB))
                Ymax2=max(BB) # After removing the maximum peak value, take the maximum value, which is the second peak value.
                Xmax1=peakposi_x[peakposi_y.index(Ymax1)] # X value corresponding to the maximum peak value
                Xmax2=peakposi_x[peakposi_y.index(Ymax2)] # X value corresponding to the second peak value
                Yvaley=[]  # Y value data set of valleys between the maximum peak and the second peak
                Xvaley=[] # X value data set of valleys between the maximum peak and the second peak
                for b in valeyposi_x:
                    if Xmax1<b<=Xmax2 or Xmax2<b<=Xmax1:
                        CC=valeyposi_y[valeyposi_x.index(b)]
                        DD=b
                        Yvaley.append(CC)
                        Xvaley.append(DD)
                # Screening conditions of bimodal distribution
                if Ymax2>=0.3*Ymax1 and (Ymax2-min(Yvaley))>=0.2*Ymax2 : 
                    posi=i
                    Bimodal_genename=dT.columns[i] 
                    Bimodal_genenames.append(Bimodal_genename)
                    Bimodal_genevalley=Xvaley[Yvaley.index(min(Yvaley))]
                    Bimodal_genevalleys.append(Bimodal_genevalley)
    # Save the results to an Excel file.
    pd.DataFrame({'Bimodal_genenames':pd.Series(Bimodal_genenames),'Bimodal_genevalleys':pd.Series(Bimodal_genevalleys)}).to_csv(del_dir2+'/'+file+'_BiDgenes(scott)',index=0,sep='\t') 
    print({file:len(Bimodal_genenames)})
```

```python
filelist=['Danish', 'Spanish', 'American','3_countries_combined', 'Human_nose','Human_cheek', 'Human_forehead', 'Pig','Mouse','Marine']
dir3='bimodal genes'
# 直接创建一个空的 DataFrame 来存储混合类型的数据
df = pd.DataFrame(columns=['bimodal gene numbers(silverman)', 'bimodal gene numbers(scott)', 'bimodal gene list(silverman)', 'bimodal gene list(scott)'], index=filelist)
for i, file in enumerate(filelist):
    silverman=pd.read_csv(dir3+'/'+filelist[i]+'_BiDgenes(silverman)',sep='\t',index_col=0)
    scott=pd.read_csv(dir3+'/'+filelist[i]+'_BiDgenes(scott)',sep='\t',index_col=0)
    bimodal_gene_list_silverman=silverman.index.tolist()
    bimodal_gene_list_scott=scott.index.tolist()
    # 将列表转换为逗号分隔的字符串
    bimodal_gene_list_silverman_str = ','.join(bimodal_gene_list_silverman)
    bimodal_gene_list_scott_str = ','.join(bimodal_gene_list_scott)
    df.loc[file, 'bimodal gene numbers(silverman)'] = len(bimodal_gene_list_silverman)
    df.loc[file, 'bimodal gene numbers(scott)'] = len(bimodal_gene_list_scott)
    df.loc[file, 'bimodal gene list(silverman)'] = bimodal_gene_list_silverman_str
    df.loc[file, 'bimodal gene list(scott)'] = bimodal_gene_list_scott_str
df.to_csv(dir3+'/bimodal gene numbers and list of silverman and scott.csv',index=True)
    
    
```
